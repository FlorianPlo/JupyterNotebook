Im folgenden geben wir eine einfache Implementierung einer (einfach) verketteten Liste (engl.: "Linked List") vor. 

* Eine verkettete Liste besteht aus einzelnen Listenelementen `ListItem`. Ein `ListItem` wiederum speichert einen Wert `value`, und eine Referenz `next` auf das nächste `ListItem`.
* Die `LinkedList` besitzt nur einen Zeiger auf das erste Element der Liste (`head`); auf die weiteren Elemente wird per `next` zugegriffen.
* Die `LinkedList` in diesem Beispiel ist Bestandteil eines Programms welches sog. Balkendiagrammen ("Barplots") erstellt. * * Ein Balkendiagramm ist hierbei definiert durch eine Menge an Balken, die als Rechtecke (`Rectangle`) visualisiert werden. Eine `LinkedList` sammelt hierbei alle Balken/ Rechtecke in ihrer Reihenfolge auf der x-Achse eines Diagramms:



````java
class Rectangle {
    int x, y, width, height;
    Rectangle(int x, int y, int width, int height) {
        this.x = x;
        this.y = y;
        this.width = width;
        this.height = height;
    }
    
    @Override
    public String toString() {
        StringBuilder sb = new StringBuilder();
        sb.append("[("+ this.x + "," + this.y + "), width: " + this.width + ", height: " + this.height + "]");
        return sb.toString();
    }
}

class ListItem {
    
    Rectangle value;
    ListItem next;
    
    ListItem(Rectangle value){
        this.value = value;
    }
}

class LinkedList {
    ListItem head;
}

/* Create a first list and its first item */
LinkedList aList = new LinkedList();
aList.head = new ListItem(new Rectangle(1, 0, 8, 4));

/* Create another item ... */
ListItem middleBar = new ListItem(new Rectangle(11, 0, 8, 10));
/* and append it to the list */
aList.head.next = middleBar;

/* Adding new items directly to the list end (via next.next and next.next.next) */
aList.head.next.next = new ListItem(new Rectangle(21, 0, 8, 29));

/* Each element in the list is only accessible by first referencing the one before */

System.out.println(aList.head.value);
System.out.println(aList.head.next.next.value);
````
